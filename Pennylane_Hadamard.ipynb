**The circuit with a single Hadamard**

```python
import pennylane as qml

# Set up PennyLane device with 1 qubit and 1 shot (for visual purposes)
dev = qml.device("default.qubit", wires=1, shots=1)

# Define the quantum function (QNode)
@qml.qnode(dev)
def coin_flip_circuit():
    # Apply Hadamard gate (H) to create equal superposition
    qml.Hadamard(wires=0)

    # Measure the qubit (sample for compatibility with Qiskit's classical bit)
    return qml.sample(wires=0)

# Draw the circuit
print(qml.draw_mpl(coin_flip_circuit)())
```

```python
import pennylane as qml
from collections import Counter
import matplotlib.pyplot as plt

# Redefine device with 10,000 shots
dev = qml.device("default.qubit", wires=1, shots=10000)

# Redefine the QNode using the updated device
@qml.qnode(dev)
def coin_flip_circuit():
    qml.Hadamard(wires=0)
    return qml.sample(wires=0)

# Run the circuit
samples = coin_flip_circuit()

# Now it will work because samples is 1D (length = 10000)
counts = Counter(samples)

# Extract counts safely
n_zeros = counts.get(0, 0)
n_ones = counts.get(1, 0)
total = n_zeros + n_ones

# Print frequencies
print("Counts:", {str(k): v for k, v in counts.items()})
print()
print("State 0 is observed with frequency %", round(100 * n_zeros / total, 2))
print("State 1 is observed with frequency %", round(100 * n_ones / total, 2))

# Plot histogram
plt.bar(["0", "1"], [n_zeros, n_ones], color=["blue", "orange"])
plt.title("Hadamard Coin Flip Result (10,000 shots)")
plt.xlabel("Measurement Outcome")
plt.ylabel("Frequency")
plt.grid(axis='y')
plt.show()
```

 The numbers of outcomes '0's and '1's are expected to be close to each other.

**As we have observed after this implementation, quantum systems output probabilistically.**

**The circuit with two Hadamards**

We design a circuit with one qubit and apply quantum coin-flipping twice

```python
import pennylane as qml

# Set up a PennyLane device with 1 qubit
dev = qml.device("default.qubit", wires=1, shots=1)

# Define the quantum circuit
@qml.qnode(dev)
def double_hadamard_circuit():
    # Apply Hadamard twice to qubit 0
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=0)

    # Return measurement sample
    return qml.sample(wires=0)

# Draw the circuit
print(qml.draw_mpl(double_hadamard_circuit)())
```

```python
# execute the circuit 10000 times in the local simulator
import pennylane as qml
from collections import Counter

# Set up PennyLane device with 1 qubit and 10,000 shots
dev = qml.device("default.qubit", wires=1, shots=10000)

# Define the circuit with two Hadamard gates
@qml.qnode(dev)
def double_hadamard():
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=0)  # H • H = I (cancels out)
    return qml.sample(wires=0)

# Run the circuit 10,000 times
samples = double_hadamard()

# Count how many times 0 and 1 were observed
counts2 = Counter(samples)

# Print the outcome counts
print("Measurement counts:", {str(k): v for k, v in counts2.items()})
```

 The only outcome must be '0'.
