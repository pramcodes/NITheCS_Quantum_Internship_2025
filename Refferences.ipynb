**One qubit gates**

x-gate (NOT operator)

```python
import pennylane as qml

# Define a device with 5 qubits
dev = qml.device("default.qubit", wires=5)

@qml.qnode(dev)
def circuit():
    qml.PauliX(wires=2)  # Apply X-gate to qubit 2
    return qml.state()
```

z-gate(Z operator)

```python
import pennylane as qml

dev = qml.device("default.qubit", wires=5)

@qml.qnode(dev)
def circuit():
    qml.PauliZ(wires=4)  # Apply Z gate to qubit 4
    return qml.state()
```

  h-gate (Hadamard operator)

```python
import pennylane as qml

dev = qml.device("default.qubit", wires=2)

@qml.qnode(dev)
def circuit():
    qml.Hadamard(wires=1)  # Apply Hadamard to qubit 1
    return qml.state()
```

  ry-gate (rotation operator)

This is a rotation operator on a (real-valued) qubit in counter-clockwise direction.

```python
import pennylane as qml
from math import pi

dev = qml.device("default.qubit", wires=4)

@qml.qnode(dev)
def circuit():
    a = pi / 7
    qml.RY(2 * a, wires=3)
    return qml.state()
```

```python
import pennylane as qml
from math import pi
a = pi / 7
dev = qml.device("default.qubit", wires=2)

@qml.qnode(dev)
def circuit():
    # Apply controlled-U3 (equivalent to CU3 in Qiskit)
    qml.ctrl(qml.Rot, control=1)(2 * a, 0, 0, wires=0)
    return qml.state()

```


**Measuring a quantum circuit**

In PennyLane, you don’t have a separate “classical register” like in Qiskit. Instead, you simply measure quantum wires, and PennyLane returns their outcomes.

1. Measuring specific wires and returning samples:

```python
import pennylane as qml

dev = qml.device("default.qubit", wires=5, shots=1000)

@qml.qnode(dev)
def circuit():
    # Add gates here (e.g., qml.X(wires=0)) if needed

    return qml.sample(wires=[0, 3, 1, 4, 2])  # Equivalent to qc.measure(q[0],c[3]), etc.
```

2. Default one-to-one mapping

```python
@qml.qnode(dev)
def circuit():
    return qml.sample(wires=range(5))  # wires 0 to 4
```

3. Partial measurement

```python
@qml.qnode(dev)
def circuit():
    return qml.sample(wires=[2, 3, 4])
```

**Emulate Conditional X Gate After Measurement**

import pennylane as qml
from pennylane import numpy as np

# Set up device with 1 qubit and 1 shot for classical behavior
dev = qml.device("default.qubit", wires=1, shots=1)

# Step 1: measure the qubit
@qml.qnode(dev)
def measure_qubit():
    qml.Hadamard(wires=0)
    return qml.sample(wires=0)  # classical 0 or 1

# Step 2: conditional X gate based on measured result
@qml.qnode(dev)
def apply_x_if_needed():
    qml.PauliX(wires=0)
    return qml.state()

@qml.qnode(dev)
def skip_x():
    return qml.state()

# Execute the logic
measurement_result = measure_qubit()[0]  # extract the single-shot result
print("Measured:", measurement_result)

# Apply conditional operation
if measurement_result == 1:
    final_state = apply_x_if_needed()
else:
    final_state = skip_x()

print("Final state:", final_state)

**Executing a quantum program on local simulator**

 Executing a quantum program 1000 times:

```python
import pennylane as qml
from collections import Counter
import numpy as np

dev = qml.device("default.qubit", wires=4, shots=1000)

@qml.qnode(dev)
def circuit():
    # Build your circuit here (e.g., qml.Hadamard(wires=0), qml.CNOT(wires=[0,1]), etc.)
    return qml.sample(wires=[0, 1, 2, 3])  # sample from all 4 wires

# Execute circuit and count measurement results
samples = circuit()  # shape: (shots, wires)
bitstrings = ["".join(str(bit) for bit in sample) for sample in samples]
counts = Counter(bitstrings)

print(counts)
```

**Reading the quantum state of a circuit on local simulator**

The quantum state of a circuit can be read when using "StatevectorSimulator".

```python
import pennylane as qml
from math import pi
import numpy as np

dev = qml.device("default.qubit", wires=4)  # specify number of qubits explicitly

@qml.qnode(dev)
def circuit():
    # example gates, replace with your circuit
    qml.PauliX(wires=0)
    qml.Hadamard(wires=1)
    qml.RY(2 * pi / 7, wires=3)
    return qml.state()

state = circuit()
precision = 3
state_rounded = np.round(state, precision)
print(state_rounded)
```

The result is a list of complex numbers. We can access the amplitude of each state by using its index, e.g.,

```python
state = circuit()  # output of qml.state(), a complex numpy array

# real part of the first amplitude
real_part = state[0].real

# imaginary part of the first amplitude
imag_part = state[0].imag

print("Real part:", real_part)
print("Imaginary part:", imag_part)
```

**Reading the unitary operator of a circuit on local simulator**

"UnitarySimulator" gives a unitary representation of all gates in the circuit until that point

```python
import pennylane as qml

# Suppose your circuit has 5 qubits (adjust wires accordingly)
dev = qml.device("default.qubit", wires=5)

@qml.qnode(dev)
def circuit():
    # apply your gates here, e.g.
    # qml.PauliX(wires=0)
    # qml.PauliX(wires=3)
    return qml.state()  # or qml.unitary() - see below

# To get the unitary matrix of the circuit, you use `qml.matrix`
unitary_matrix = qml.matrix(circuit)()

print(unitary_matrix.round(3))  # round for precision as in your code
```

The result is a double list of complex numbers (as it is a matrix).

```python
real_part = unitary_matrix[0, 0].real
imag_part = unitary_matrix[0, 0].imag

print("Real part:", real_part)
print("Imag part:", imag_part)
```
