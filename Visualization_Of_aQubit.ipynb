Suppose that we have a single qubit.

Each possible (real-valued) quantum state of this qubit is a point on 2-dimensional space.

It can also be represented as a vector from origin to that point.

We start with the visual representation of the following quantum states:


    $$ \ket{0} =  \myvector{1\\0}, ~~ \ket{1} = \myvector{0\\1} , ~~ -\ket{0} = \myrvector{-1\\0}, \mbox{and} -\ket{1} = \myrvector{0\\-1}. $$
    
We draw these quantum states as points.

We use one of our predefined functions for drawing axes: "draw_axes()". We include our predefined functions with the following line of code:

```python
import matplotlib.pyplot as plt

# Custom axis drawing (replaces draw_axes())
def draw_axes():
    plt.axhline(0, color='gray', linestyle='--', linewidth=1)
    plt.axvline(0, color='gray', linestyle='--', linewidth=1)
    plt.xlim(-1.2, 1.2)
    plt.ylim(-1.2, 1.2)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(True)

# Create figure
plt.figure(figsize=(6, 6), dpi=80)

# Draw axes
draw_axes()

# Draw origin in red
plt.plot(0, 0, 'ro')

# Plot quantum state projections
plt.plot(1, 0, 'bo')   # |+> (X-axis)
plt.plot(0, 1, 'go')   # |i> (Y-axis)
plt.plot(-1, 0, 'yo')  # |-> (-X-axis)
plt.plot(0, -1, 'co')  # |-i> (-Y-axis)

# Show plot
plt.title("2D Projection of Quantum States (Bloch XY Plane)")
plt.show()
```

 Now, we draw the quantum states as arrows (vectors):

```python
import matplotlib.pyplot as plt

# Draw axes (replacement for draw_axes())
def draw_axes():
    plt.axhline(0, color='gray', linestyle='--', linewidth=1)
    plt.axvline(0, color='gray', linestyle='--', linewidth=1)
    plt.xlim(-1.2, 1.2)
    plt.ylim(-1.2, 1.2)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(True)

# Create figure
plt.figure(figsize=(6, 6), dpi=80)

# Draw axes
draw_axes()

# Draw vectors (quantum states)
plt.arrow(0, 0, 0.92, 0, head_width=0.04, head_length=0.08, color="r", label='|+⟩')   # X+
plt.arrow(0, 0, 0, 0.92, head_width=0.04, head_length=0.08, color="b", label='|i⟩')   # Y+
plt.arrow(0, 0, -0.92, 0, head_width=0.04, head_length=0.08, color="g", label='|−⟩')  # X−
plt.arrow(0, 0, 0, -0.92, head_width=0.04, head_length=0.08, color="y", label='|−i⟩') # Y−

plt.title("Quantum States as Vectors (XY Projection of Bloch Sphere)")
plt.show()
```

# Unit Circle

All (real-valued) quantum states of a qubit form the unit circle.

The length of each quantum state is 1.

All points that are 1 unit away from the origin form the circle with radius 1 unit.

We can draw the unit circle with python.

We have a predefined function for drawing the unit circle: "draw_unit_circle()"

```python
import matplotlib.pyplot as plt

# Define the figure
plt.figure(figsize=(6, 6), dpi=80)

# Draw axes
plt.axhline(0, color='gray', linestyle='--', linewidth=1)
plt.axvline(0, color='gray', linestyle='--', linewidth=1)

# Draw unit circle
circle = plt.Circle((0, 0), radius=1, fill=False, color='black', linestyle='-')
plt.gca().add_patch(circle)

# Set aspect and limits
plt.gca().set_aspect('equal', adjustable='box')
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.grid(True)

# Show the plot
plt.title("2D Bloch Sphere Projection with Unit Circle")
plt.show()
```

# Quantum state of a qubit

 Suppose that we have a single qubit.

Each possible (real-valued) quantum state of this qubit is a point on 2-dimensional space.

It can also be represented as a vector from origin to that point.We draw the quantum state
    
    $ \myvector{3/5 \\ 4/5} $

and its elements.

*Our predefined function "draw_qubit()" draws a figure, the origin, the axes, the unit circle, and base quantum states.*

*Our predefined function "draw_quantum_state(x,y,name)" draws an arrow from (0,0) to (x,y) and associates it with name.*

*We include our predefined functions with the following line of code.*

```python
import matplotlib.pyplot as plt

def draw_qubit():
    plt.figure(figsize=(6, 6), dpi=80)
    plt.axhline(0, color='gray', linestyle='--')
    plt.axvline(0, color='gray', linestyle='--')
    circle = plt.Circle((0, 0), radius=1, fill=False, color='black')
    plt.gca().add_patch(circle)
    plt.xlim(-1.2, 1.2)
    plt.ylim(-1.2, 1.2)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(True)

def draw_quantum_state(x, y, label="|ψ⟩"):
    plt.arrow(0, 0, x, y, head_width=0.04, head_length=0.06, color='blue')
    plt.text(x + 0.05, y + 0.05, label, fontsize=12, color='blue')

def show_plt():
    plt.title("Quantum State |v⟩ on 2D Bloch Sphere Projection")
    plt.show()

# Now call the same logic
draw_qubit()
draw_quantum_state(3/5, 4/5, "|v⟩")
show_plt()
```

 Now, we draw its angle with ∣0⟩-axis and its projections on both axes.

 For drawing the angle, we use the method "Arc" from library "matplotlib.patches"

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc

def draw_qubit():
    plt.figure(figsize=(6, 6), dpi=80)
    plt.axhline(0, color='gray', linestyle='--')
    plt.axvline(0, color='gray', linestyle='--')
    circle = plt.Circle((0, 0), radius=1, fill=False, color='black')
    plt.gca().add_patch(circle)
    plt.xlim(-1.2, 1.2)
    plt.ylim(-1.2, 1.2)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(True)

def draw_quantum_state(x, y, label="|ψ⟩"):
    plt.arrow(0, 0, x, y, head_width=0.04, head_length=0.06, color='blue')
    plt.text(x + 0.05, y + 0.05, label, fontsize=12, color='blue')

# Values
x = 3/5
y = 4/5

# Plot
draw_qubit()
draw_quantum_state(x, y, "|v⟩")

# Projection on |0⟩ axis
plt.arrow(0, 0, x, 0, color="blue", linewidth=1.5)
plt.arrow(0, y, x, 0, color="blue", linestyle='dotted')
plt.text(0.1, -0.1, "cos(a)=3/5", fontsize=10)

# Projection on |1⟩ axis
plt.arrow(0, 0, 0, y, color="blue", linewidth=1.5)
plt.arrow(x, 0, 0, y, color="blue", linestyle='dotted')
plt.text(-0.1, 0.55, "sin(a)=4/5", rotation=90, fontsize=10)

# Draw the angle arc
plt.gca().add_patch(Arc((0, 0), 0.4, 0.4, angle=0, theta1=0, theta2=53))
plt.text(0.08, 0.05, '.', fontsize=30)
plt.text(0.21, 0.09, 'a', fontsize=12)

# Final touches
plt.title("Quantum State |v⟩ with Projections and Angle")
plt.show()
```

 ** Observations:**

 * The angle of quantum state with state

* The amplitude of state
    
    
    $ \\ket{1} $

  **The angle of a quantum state**

  The angle of a vector (in radians) on the unit circle is the length of arc in counter-clockwise direction that starts from (1,0) and with the points representing the vector.
  
  We execute the following code a couple of times to see different examples, where the angle is picked randomly in each run.
  
  You can also set the value of "myangle" manually for seeing a specific angle

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Arc
from math import sin, cos, pi
from random import randrange

# Set random angle
myangle = randrange(361)

# Convert angle to radians
myangle_in_radian = 2 * pi * (myangle / 360)
x = cos(myangle_in_radian)
y = sin(myangle_in_radian)

# Calculated values
ratio_of_arc = ((1000 * myangle / 360) // 1) / 1000

print(f"the selected angle is {myangle} degrees")
print(f"it is {ratio_of_arc} of a full circle")
print(f"its length is {ratio_of_arc} x 2π = {ratio_of_arc * 2 * pi}")
print(f"its radian value is {myangle_in_radian}")
print()
print(f"the amplitude of state |0> is {x}")
print(f"the amplitude of state |1> is {y}")
print()
print(f"the probability of observing state |0> is {x**2}")
print(f"the probability of observing state |1> is {y**2}")
print(f"the total probability is {round(x**2 + y**2, 6)}")

# Setup plot
plt.figure(figsize=(6, 6), dpi=60)
ax = plt.gca()
ax.set_aspect('equal')
plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.grid(True)
plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')

# Draw unit circle
circle = plt.Circle((0, 0), radius=1, fill=False, color='black')
ax.add_patch(circle)

# Small red arc showing angle
ax.add_patch(Arc((0, 0), 0.2, 0.2, angle=0, theta1=0, theta2=myangle, color="red", linewidth=2))

# Full-circle arc in brown
ax.add_patch(Arc((0, 0), 2, 2, angle=0, theta1=0, theta2=myangle, color="brown", linewidth=2))

# Quantum state vector
plt.arrow(0, 0, x, y, head_width=0.04, head_length=0.06, color='blue')
plt.text(x + 0.05, y + 0.05, "|v⟩", fontsize=12, color='blue')

# Projection on |0> axis (x-axis)
plt.arrow(0, 0, x, 0, color="blue", linewidth=1)
plt.arrow(0, y, x, 0, color="blue", linestyle='dashed')

# Projection on |1> axis (y-axis)
plt.arrow(0, 0, 0, y, color="blue", linewidth=1)
plt.arrow(x, 0, 0, y, color="blue", linestyle='dashed')

# Show plot
plt.title("Quantum State |v⟩ at Random Angle on Bloch Sphere Projection")
plt.show()
```
