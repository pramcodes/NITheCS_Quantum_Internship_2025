Quantum systems are linear systems: "The quantum states are represented by vectors and quantum operators are represented by matrices. The new quantum states are calculated by corresponding matrix-vector multiplications."

A qubit (quantum bit) has two states: state 0 and state 1.

They are denoted by ket-notation:

|0> = [1,0] and |1> = [0,1]

**NOT operator**
NOT operator flips the value of a qubit.

We use capital letter for the matrix form of the operators:

    $ X = \X$

The action of X on the qubit:

    $ X \ket{0} = \ket{1} $.

 More explicitly,
    
    $ X \ket{0} = \X \vzero = \vone = \ket{1} $.

 Similarly, X∣1⟩=∣0⟩.More explicitly,


     $ X \ket{1} = \X \vone = \vzero = \ket{0} $.

**Hadamard operator**

Hadamard operator ($ H $ or h-gate) looks similar to a fair coin-flipping.


    $$ H = \hadamard.
    $$

But, there are certain dissimilarities:

     

     we have a negative entry, and

     instead of $ \\frac{1}{2} $, we have its square root $ \\mypar{ \\frac{1}{\\sqrt{2}} } $.


 Quantum systems can have negative transitions.  

 Can probabilistic systems be extended with negative values ?

**One-step Hadamard**

Start in ∣0⟩.

After applying H:


    $$ H \ket{0} =  \hadamard \vzero =  \stateplus.
    $$

After measurement, we observe the states ∣0⟩ and ∣1⟩ with equal probability 1/2​. How can this be possible when their values are 1/2​?

Let's start in ∣1⟩.After applying H:

    $$ H \ket{1} = \hadamard \vone = \stateminus.
    $$

After measurement, we observe the states ∣0⟩ and ∣1⟩ with equal probability 1/2​.

We obtain the same values even when one of the values is negative


 The absolute value of a negative value is positive.The square of a negative value is also positive.As we have observed, the second fact fits better when reading the measurement results.

**When a quantum system is measured, the probability of observing one state is the square of its value [*].**

The value of the system being in a state is called its amplitude.

In the above example, the amplitudes of states

 ∣0⟩ and ∣1⟩ are respectively


    $ \sqrttwo $ and $ -\sqrttwo $.

 The probabilities of observing them after a measurement are

    $ \onehalf $.
    
 *Remark that, after observing state 0, the new state will be ∣0⟩, and, after observing state 1, the new state will be ∣1⟩.*

  
  [*] the square of its magnitude value when using complex numbers

    **States $ \ket{+} $ and $ \ket{-}**
   In the literature, the quantum states obtained after applying H to ∣0⟩ and ∣1⟩ are known as ket-plus ($\ket{+}$) and ket-minus ($ \ket{-} $) states, respectively:

    $ \ket{+} = H \ket{0} = \hadamard \vzero = \stateplus $
   
    $ \ket{-} = H \ket{1} =  \hadamard \vone = \stateminus $
