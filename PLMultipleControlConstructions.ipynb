<h3>Solution</h3>

#### Implementation of $ C_0C_0NOT $ operator

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import UnitarySimulator

q =  QuantumRegister(3,"q")
c = ClassicalRegister(3,"c")

qc = QuantumCircuit(q,c)

qc.x(q[2])
qc.x(q[1])
qc.ccx(q[2],q[1],q[0])
qc.x(q[2])
qc.x(q[1])

job = UnitarySimulator().run(qc, shots = 1)
U=job.result().get_unitary(qc,decimals=3).data

print("CCNOT(00) = ")
for row in U:
    s = ""
    for value in row:
        s = s + str(round(value.real,2)) + "  "
    print(s)



qc.draw(output="mpl",reverse_bits=True)
```

```python
import pennylane as qml
import numpy as np
import matplotlib.pyplot as plt


# Create a 2-qubit quantum device
dev = qml.device("default.qubit", wires=3)

@qml.qnode(dev)
def circuit(state):
    qml.Toffoli(wires=[0,1,2])
    return qml.state()

state=np.array([1,0])
matrix = qml.matrix(circuit)(state)
print("CCNOT(00) = ", matrix)

# Define the quantum circuit with classical inputs a and b
@qml.qnode(dev)
def quantum_circuit():
    qml.PauliX(wires=0)
    qml.PauliX(wires=1)
    qml.Toffoli(wires=[0,1,2])
    qml.PauliX(wires=0)
    qml.PauliX(wires=1)
    return qml.state()

fig, ax = qml.draw_mpl(quantum_circuit, style="black_white")()
plt.show()



```

Provide an implementation of the NOT operator controlled by 4 qubits ($CCCCNOT$) in Qiskit. Verify its correctness by executing your solution on all possible inputs. (See the above example)

*You may use two auxiliary qubits.*

<h3>Solution</h3>

#### Desinging the operator

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import UnitarySimulator

qaux = QuantumRegister(2,"qaux")
q = QuantumRegister(5,"q")
c = ClassicalRegister(5,"c")
qc = QuantumCircuit(q,qaux,c)

qc.ccx(q[4],q[3],qaux[1])
qc.ccx(q[2],q[1],qaux[0])
qc.ccx(qaux[1],qaux[0],q[0])
qc.ccx(q[4],q[3],qaux[1])
qc.ccx(q[2],q[1],qaux[0])

qc.draw(output="mpl",reverse_bits=True)
```

```python

dev = qml.device("default.qubit", wires=7)

@qml.qnode(dev)
def circuit():

    qml.Toffoli(wires=[3,2,0])
    qml.Toffoli(wires=[5,4,1])
    qml.Toffoli(wires=[0,1,6])
    qml.Toffoli(wires=[3,2,0])
    qml.Toffoli(wires=[5,4,1])
    return qml.state()


fig, ax = qml.draw_mpl(circuit, style="black_white")()
plt.show()
```

#### Executing this circuit on every possible inputs and verify the correctness of the implementation experimentially.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator

all_inputs=[]
for q4 in ['0','1']: 
    for q3 in ['0','1']:
        for q2 in ['0','1']:
            for q1 in ['0','1']:
                for q0 in ['0','1']:
                    all_inputs.append(q4+q3+q2+q1+q0)
#print(all_inputs)  

print("input --> output")
for the_input in all_inputs:
    # create the circuit
    qaux = QuantumRegister(2,"qaux")
    q = QuantumRegister(5,"q")
    c = ClassicalRegister(5,"c")
    qc = QuantumCircuit(q,qaux,c)
    # set the initial value of the circuit w.r.t. the input
    if the_input[0] =='1': qc.x(q[4])
    if the_input[1] =='1': qc.x(q[3])
    if the_input[2] =='1': qc.x(q[2])
    if the_input[3] =='1': qc.x(q[1])
    if the_input[4] =='1': qc.x(q[0])
    # implement the CCNOT gates
    qc.ccx(q[4],q[3],qaux[1])
    qc.ccx(q[2],q[1],qaux[0])
    qc.ccx(qaux[1],qaux[0],q[0])
    qc.ccx(q[4],q[3],qaux[1])
    qc.ccx(q[2],q[1],qaux[0])
    # measure the main quantum register 
    qc.measure(q,c)
    # execute the circuit
    job = AerSimulator().run(qc,shots=1)
    counts = job.result().get_counts(qc)
    for key in counts: the_output = key
    printed_str = the_input[0:4]+" "+the_input[4]+" --> "+the_output[0:4]+" "+the_output[4]
    if (the_input!=the_output): printed_str = printed_str + " the output is different than the input"
    print(printed_str)
```

```python

import pennylane as qml
import numpy as np

# Device with 7 qubits: 5 main + 2 ancilla
dev = qml.device("default.qubit", wires=7, shots=1)

# Function to build and run the circuit
def run_circuit(input_bits):
    @qml.qnode(dev)
    def circuit():
        # Initialize input bits on wires [4,3,2,1,0]
        if input_bits[0] == '1':
            qml.PauliX(wires=4)
        if input_bits[1] == '1':
            qml.PauliX(wires=3)
        if input_bits[2] == '1':
            qml.PauliX(wires=2)
        if input_bits[3] == '1':
            qml.PauliX(wires=1)
        if input_bits[4] == '1':
            qml.PauliX(wires=0)

        # Apply Toffoli gates (ancilla: 5 and 6)
        qml.Toffoli(wires=[4, 3, 6])   # q[4], q[3] -> qaux[1] (wire 6)
        qml.Toffoli(wires=[2, 1, 5])   # q[2], q[1] -> qaux[0] (wire 5)
        qml.Toffoli(wires=[6, 5, 0])   # qaux[1], qaux[0] -> q[0]
        qml.Toffoli(wires=[4, 3, 6])   # uncompute ancilla
        qml.Toffoli(wires=[2, 1, 5])   # uncompute ancilla

        # Return a sample of the output state (wires 4→0 = original q[4:0])
        return qml.sample(wires=[4,3,2,1,0])
    
    return circuit()

# Loop over all possible 5-bit inputs
all_inputs = [f"{i:05b}" for i in range(32)]
print("input --> output")
for input_bits in all_inputs:
    sample = run_circuit(input_bits)
    output_bits = ''.join(str(bit) for bit in sample)
    printed_str = f"{input_bits[:4]} {input_bits[4]} --> {output_bits[:4]} {output_bits[4]}"
    if input_bits != output_bits:
        printed_str += " the output is different than the input"
    print(printed_str)
```

```python
import pennylane as qml
import numpy as np

# Device with 7 qubits: q[0]–q[4] (main), qaux[0]–qaux[1] (ancilla = wires 5,6)
dev = qml.device("default.qubit", wires=7, shots=1)

# Function to build and run the circuit for a given input string
def run_circuit(the_input):
    @qml.qnode(dev)
    def circuit():
        # Initialize the qubits q[4] to q[0] from the input bits
        if the_input[0] == '1': qml.PauliX(wires=4)
        if the_input[1] == '1': qml.PauliX(wires=3)
        if the_input[2] == '1': qml.PauliX(wires=2)
        if the_input[3] == '1': qml.PauliX(wires=1)
        if the_input[4] == '1': qml.PauliX(wires=0)

        # Extra X gates (inverting controls on wires 3 and 1)
        qml.PauliX(wires=3)
        qml.PauliX(wires=1)

        # Toffoli sequence with ancilla
        qml.Toffoli(wires=[4, 3, 6])  # q[4], q[3] → qaux[1]
        qml.Toffoli(wires=[2, 1, 5])  # q[2], q[1] → qaux[0]
        qml.Toffoli(wires=[6, 5, 0])  # qaux[1], qaux[0] → q[0]
        qml.Toffoli(wires=[4, 3, 6])  # uncompute
        qml.Toffoli(wires=[2, 1, 5])  # uncompute

        # Undo extra X gates
        qml.PauliX(wires=3)
        qml.PauliX(wires=1)

        # Measure q[4:0]
        return qml.sample(wires=[4, 3, 2, 1, 0])

    return circuit()

# Run for all possible 5-bit input combinations
all_inputs = [f"{i:05b}" for i in range(32)]

print("input --> output")
for the_input in all_inputs:
    sample = run_circuit(the_input)
    the_output = ''.join(str(bit) for bit in sample)

    printed_str = f"{the_input[0:4]} {the_input[4]} --> {the_output[0:4]} {the_output[4]}"
    if the_input != the_output:
        printed_str += " the output is different than the input"

    print(printed_str)
```

<h3>Solution</h3>

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator

all_inputs=[]
for q4 in ['0','1']: 
    for q3 in ['0','1']:
        for q2 in ['0','1']:
            for q1 in ['0','1']:
                for q0 in ['0','1']:
                    all_inputs.append(q4+q3+q2+q1+q0)
#print(all_inputs)  


def c4not(control_state='1111'):
    #
    # drawing the circuit
    #
    print("Control state is",control_state)
    print("Drawing the circuit:")
    qaux = QuantumRegister(2,"qaux")
    q = QuantumRegister(5,"q")
    c = ClassicalRegister(5,"c")
    qc = QuantumCircuit(q,qaux,c)

    for b in range(4):
        if control_state[b] == '0':
            qc.x(q[4-b])
    qc.ccx(q[4],q[3],qaux[1])
    qc.ccx(q[2],q[1],qaux[0])
    qc.ccx(qaux[1],qaux[0],q[0])
    qc.ccx(q[4],q[3],qaux[1])
    qc.ccx(q[2],q[1],qaux[0])
    for b in range(4):
        if control_state[b] == '0':
            qc.x(q[4-b])
            
    qc.draw(output="text", reverse_bits=True)
    
    #
    # executing the operator on all possible inputs
    #
    print("Control state is",control_state)
    print("input --> output")
    for the_input in all_inputs:
        # create the circuit
        qaux = QuantumRegister(2,"qaux")
        q = QuantumRegister(5,"q")
        c = ClassicalRegister(5,"c")
        qc = QuantumCircuit(q,qaux,c)
        # set the initial value of the circuit w.r.t. the input
        if the_input[0] =='1': qc.x(q[4])
        if the_input[1] =='1': qc.x(q[3])
        if the_input[2] =='1': qc.x(q[2])
        if the_input[3] =='1': qc.x(q[1])
        if the_input[4] =='1': qc.x(q[0])
        # implement the CCNOT gates
        for b in range(4):
            if control_state[b] == '0':
                qc.x(q[4-b])
        qc.ccx(q[4],q[3],qaux[1])
        qc.ccx(q[2],q[1],qaux[0])
        qc.ccx(qaux[1],qaux[0],q[0])
        qc.ccx(q[4],q[3],qaux[1])
        qc.ccx(q[2],q[1],qaux[0])
        for b in range(4):
            if control_state[b] == '0':
                qc.x(q[4-b])
        # measure the main quantum register 
        qc.measure(q,c)
        # execute the circuit
        job = AerSimulator().run(qc,shots=1)
        counts = job.result().get_counts(qc)
        for key in counts: the_output = key[::-1] 
        printed_str = the_input[0:4]+" "+the_input[4]+" --> "+the_output[0:4]+" "+the_output[4]
        if (the_input!=the_output): printed_str = printed_str + " the output is different than the input"
        print(printed_str)
```
