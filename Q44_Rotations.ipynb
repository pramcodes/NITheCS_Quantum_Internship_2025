<h3> Rotation with degree $\pi/4$ </h3>

Its matrix form is similar to Hadamard operator:

$$ R(\pi/4) = \mymatrix{rr}{\cos(\pi/4) & -\sin(\pi/4) \\ \sin(\pi/4) & \cos(\pi/4) }
= \mymatrix{rr}{ \sqrttwo & -\sqrttwo \\  \sqrttwo & \sqrttwo} $$.

<h3> Task 2 </h3>

Start with state $ \ket{0} $. 

Apply $ R(\pi/4) $ 7 times, and draw each state on the unit circle.

```python
%run quantum.py

draw_qubit()

[x,y]=[1,0]

draw_quantum_state(x,y,"v0")

sqrttwo = 2**0.5
oversqrttwo = 1/sqrttwo

R = [ [oversqrttwo, -1*oversqrttwo], [oversqrttwo,oversqrttwo] ]

# function for rotation R
def rotate(px,py):
    newx = R[0][0]*px + R[0][1]*py
    newy = R[1][0]*px + R[1][1]*py
    return [newx,newy]

# apply rotation R 7 times
for i in range(1,8):
    [x,y] = rotate(x,y)
    draw_quantum_state(x,y,"|v"+str(i)+">")

show_plt()
```

<h3>Reading the quantum state of a circuit</h3>

PennyLane provides access to the full quantum state through its classical simulators. In particular, the "default.qubit" device can simulate and return the state vector of the quantum circuit at the end of execution.

To read the quantum state of a circuit in PennyLane, we use the qml.state() return type inside a QNode. This gives direct access to the list of complex amplitudes representing the quantum state.

<h3> Task 3 </h3>

We implement Task 2 by using "ry-gate" and "statevector_simulator".

Define a quantum circuit with one qubit. 

Iterate seven times
- Apply the rotation with angle $ \pi \over 4 $ by using ry-gate
- Read the current quantum state
- Print the current quantum state
- Draw the quantum state on the unit circle by using the values of the current quantum state

```python


import pennylane as qml
import numpy as np
from math import pi
from quantum import draw_qubit, draw_quantum_state, show_plt  # assuming quantum.py is available

# Start with drawing the initial state vector
draw_qubit()
[x, y] = [1, 0]
draw_quantum_state(x, y, "v0")

# Create a device that gives full statevector access
dev = qml.device("default.qubit", wires=1)

# Define a QNode with a parameter for cumulative rotation
@qml.qnode(dev)
def rotation_circuit(theta):
    qml.RY(theta, wires=0)
    return qml.state()

# Accumulate rotations and draw state after each one
rotation_angle = pi / 4
cumulative_theta = 0.0

for i in range(1, 8):
    cumulative_theta += 2 * rotation_angle  # same as R(pi/2)
    
    # Get the full quantum state vector
    state = rotation_circuit(cumulative_theta)
    
    x_val = state[0].real  # amplitude of |0>
    y_val = state[1].real  # amplitude of |1>
    
    print(f"iteration {i}: the quantum state is ({x_val:.4f}) |0> + ({y_val:.4f}) |1>")
    draw_quantum_state(x_val, y_val, f"|v{i}>")

show_plt()
```

<h3> Rotation with degree $\theta$ </h3>

The matrix form of a rotation is as follows:

$$ R(\theta) = \mymatrix{rr}{\cos(\theta) & -\sin(\theta) \\ \sin(\theta) & \cos(\theta) }, $$

where $ \theta $ is the angle of rotation (in counter-clockwise direction).

<h3> Task 6 </h3>

We randomly pick an angle $ \theta \in [0,2\pi) $.

We have two separate qubits initially set to state $ \ket{0} $. 

The first qubit is rotated by the angle $ \theta $ and the second qubit is rotated by the angle $ \theta + \pi/2 $.

Implement each qubit and its rotation as a separate quantum circuit.

Then, read both quantum states and calculate their dot product.

Check the result of the dot product for different random angles.

```python
import pennylane as qml
from random import randrange
from math import pi
import numpy as np

# Device with 1 qubit
dev = qml.device("default.qubit", wires=1)

# QNode for rotating a qubit
@qml.qnode(dev)
def rotation_circuit(theta):
    qml.RY(theta, wires=0)
    return qml.state()

# Perform the experiment 10 times
for i in range(10):
    # Pick a random angle
    random_angle = randrange(3600) / 10

    # Specify the rotation angles
    rotation_angle1 = (random_angle / 360) * 2 * pi
    rotation_angle2 = rotation_angle1 + pi / 2

    # Run first circuit
    state1 = rotation_circuit(2 * rotation_angle1)
    x1, y1 = state1[0].real, state1[1].real

    # Run second circuit
    state2 = rotation_circuit(2 * rotation_angle2)
    x2, y2 = state2[0].real, state2[1].real

    # Compute and print dot product
    dot_product = round(x1 * x2 + y1 * y2, 5)
    print(f"{i} - the result of dot product is {dot_product}")
    print(f"random angle is {random_angle}")
    print(f"x1 , y1 = {round(x1, 5)} {round(y1, 5)}")
    print(f"x2 , y2 = {round(x2, 5)} {round(y2, 5)}\n")

```

<h3> Task 7 </h3>

We randomly pick an angle $ \theta \in [0,2\pi) $.

We have a single qubit initially set to state $ \ket{0} $. 

The qubit is rotated by the angle either $ \theta_1 =  \theta $ or $ \theta_2 =  \theta-\pi/2 $.

You are allowed to do one more rotation $ \theta' $ and then make a measurement.

Can you determine the angle of the first rotation angle by looking/using the measurement result? Is it $ \theta_1 $ or $ \theta_2 $?

Check your solution for different random angles.

```python
import pennylane as qml
from random import randrange
from math import pi
import numpy as np

# Step 1: Setup angles
random_angle = randrange(3600) / 10
rotation_angle1 = (random_angle / 360) * 2 * pi
rotation_angle2 = rotation_angle1 - pi / 2

# Step 2: Randomly apply theta1 or theta2
picked_angle = "theta1" if randrange(2) == 0 else "theta2"

# Step 3: PennyLane device (shots to simulate measurements)
dev = qml.device("default.qubit", wires=1, shots=100)

# Step 4: Quantum node to apply rotations and measure
@qml.qnode(dev)
def circuit():
    if picked_angle == "theta1":
        qml.RY(2 * rotation_angle1, wires=0)
    else:
        qml.RY(2 * rotation_angle2, wires=0)
    
    # Undo theta1 rotation
    qml.RY(-2 * rotation_angle1, wires=0)

    # Measure in computational basis
    return qml.sample(qml.PauliZ(0))

# Step 5: Run and interpret results
samples = circuit()
counts = {
    '0': np.sum(samples == 1),  # |0 corresponds to +1
    '1': np.sum(samples == -1)  # |1 corresponds to -1
}

# Step 6: Guess based on result
if counts['0'] > counts['1']:
    your_guess = "theta1"
else:
    your_guess = "theta2"

# Final Output
print("Measurement counts:", counts)
print("your guess is", your_guess)
print("picked_angle is", picked_angle)
```
